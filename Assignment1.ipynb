```python
# imports
import os
import pandas as panda 
from matplotlib import pyplot as plot
```

```python
dataframe = panda.read_csv("Ruter_data.csv", sep=';', error_bad_lines=False)
dataframe.head()
```

```python
dataframe.shape
```

```python
dataframe.sample(10)
```

```python
# Passenger distrubution with Histogram
plot.hist(dataframe.Passasjerer_Ombord, bins=25, rwidth=1)
plot.xlabel('Passengers Onboard')
plot.ylabel('Count')
plot.show()
```

```python
#Passenger column with mean and standard deviation
dataframe.Passasjerer_Ombord.describe()
```

```python
#Date column 
dataframe.Dato.describe()
```

```python
#Municipality column
dataframe.Kommune.describe()
```

```python
#Line name columnKjøretøy_Kapasitet
dataframe.Linjenavn.describe()
```

```python
#Capacity column
dataframe.Kjøretøy_Kapasitet.describe()
```

```python
#detect outliers
passenger_lower_limit, passenger_upper_limit = dataframe.Passasjerer_Ombord.quantile([0.02,0.998])
capacity_lower_limit, capacity_upper_limit = dataframe.Kjøretøy_Kapasitet.quantile([0.01,0.998])
passenger_lower_limit, passenger_upper_limit
```

```python
dataframe[dataframe.Passasjerer_Ombord<passenger_lower_limit]
```

```python
dataframe[dataframe.Passasjerer_Ombord>passenger_upper_limit]
```

```python
new_dataframe = dataframe[(dataframe.Passasjerer_Ombord>passenger_lower_limit)&(dataframe.Passasjerer_Ombord<passenger_upper_limit)]
new_dataframe = dataframe[(dataframe.Kjøretøy_Kapasitet>capacity_lower_limit)&(dataframe.Kjøretøy_Kapasitet<capacity_upper_limit)]
new_dataframe.shape
```

```python
# New Passenger distrubution with Histogram
plot.hist(new_dataframe.Passasjerer_Ombord, bins=25, rwidth=1)
plot.xlabel('Passengers Onboard')
plot.ylabel('Count')
plot.show()
```

```python
new_dataframe[100:300].drop(['Kommune'], axis=1).plot.line(title='Ruter Dataset')
```
